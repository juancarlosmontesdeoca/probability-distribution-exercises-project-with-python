# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

DISTRIBUCION NORMAL

Funciones para la Distribucion normal

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
```

Funcion de densidad pdf

```python
def dens_normal(x, loc, scale):
  result = norm.pdf(x, loc, scale)
  return result
```

Funcion de Probabilidad Acumulada

```python
def acum_normal(x, loc, scale):
  result = norm.cdf(x, loc, scale)
  return result
```

Funcion Percentil

```python
def percen_normal(x, loc, scale):
  result = norm.ppf(x, loc, scale)
  return result
```

Generacion de numeros aleatorios bajo una distribucion normal

```python
def nros_random_normal(n, size, prob):
  result = norm.rvs(n, size, prob)
  return result
```

```python
loc = 23
scale = 2.75
x = 23
q = 0.75
n = 20
prob = 0.35
size = 10

dens_normal(x, loc, scale)
acum_normal(x, loc, scale)
percen_normal(q, loc, scale)
nros_random_normal(n, loc, size)
print()
```

```python
print(f"Valores asociados a una Distribucion Normal de media {loc} y desviacion estandar {scale}")
print()
print(f"Probabilidad que que sea menor a {x} : {dens_normal(x, loc, scale)}")
print()
print(f"Probabilidad acumulada de los valores menores a {x}: {acum_normal(x, loc, scale)}")
print()
print(f"Datos que han acumulado el Percentil {str(q*100)+'%'}, {percen_normal(q, loc, scale)}")
print()
print(f"Generacion de {size} numeros aleatorios:\n {nros_random_normal(n, loc, size)}")
```

```python

#Establecemos los parametros de la Normal mencionados arriba
loc = 23
scale = 2.75

#Creamos el eje x con la media (loc) y 3 veces la desviacion estandar donde se concentra el 99,7% de los datos
x = np.linspace(loc-3*scale, loc+3*scale, 100)

#Definimos las funciones de densidad y acumulada
pdf = dens_normal(x, loc, scale)
cdf = acum_normal(x, loc, scale)

#Definimos la funcion percentil
ppf = percen_normal(0.75, loc, scale)

#Eje x para ppf (funcion percentil) por ser una funcion continua
q = np.linspace(0.01, 0.99, 100)
ppf_curva = percen_normal(q, loc, scale)

#Creando graficas

fig, axis = plt.subplots(3, 1, figsize=(6, 8))

#Grafica pdf - Densidad de Probabilidad
axis[0].plot(x, pdf, "r-")
axis[0].set_title("PDF - Densidad de Probabilidad - Distribucion Normal")

#Grafica cdf - Distribucion Acumulada
axis[1].plot(x, cdf, "b-")
axis[1].set_title("CDF - Distribución Acumulada - Distribucion Normal")

#Grafica ppf - Funcion Percentil
axis[2].plot(q, ppf_curva, "g-")
axis[2].set_title("PPF - Funcion Percentil - Distribucion Normal")
axis[2].axvline(0.75, color="blue", linestyle="--")
axis[2].axhline(ppf, color="gray", linestyle="--")
axis[2].plot(0.75, ppf, "go", label="Percentil 75%")
axis[2].legend()

plt.tight_layout()
plt.show()
print()
print(f'El valor donde se encuentra el 75% de los datos es {ppf}')
```

DISTRIBUCION UNIFORME

Funcion de densidad pdf

```python
def dens_uniforme(x, minimo, maximo):
  result = uniform.pdf(x, minimo, maximo)
  return result
```

Funcion de Probabilidad Acumulada

```python
def acum_uniforme(x, minimo, maximo):
  result = uniform.cdf(x, minimo, maximo)
  return result
```

Funcion Percentil

```python
def percen_uniforme(x, minimo, maximo):
  result = uniform.ppf(x, minimo, maximo)
  return result
```

Generacion de numeros aleatorios bajo una distribucion uniforme

```python
def nros_random_uniforme(minimo, amplitud, size):
  result = uniform.rvs(minimo, maximo, size)
  return result
```

```python
minimo = 0
maximo = 10
amplitud = maximo - minimo
x = 5
q = 0.50
size = 10

dens_uniforme(x, minimo, amplitud)
acum_uniforme(x, minimo, amplitud)
percen_uniforme(q, minimo, amplitud)
nros_random_uniforme(minimo, amplitud, size)
print()
```

```python
print(f"Valores asociados a una Distribucion Uniforme con minimo {minimo} y maximo {amplitud}:")
print()
print(f"Probabilidad que que sea menor a {x} : {dens_uniforme(x, minimo, amplitud)}")
print()
print(f"Probabilidad acumulada de los valores menores a {x}: {acum_uniforme(x, minimo, amplitud)}")
print()
print(f"Datos que han acumulado el Percentil {str(q*100)+'%'}, {percen_uniforme(q, minimo, amplitud)}")
print()
print(f"Generacion de {size} numeros aleatorios con Distribucion Uniforme [0,10]:\n {nros_random_uniforme(minimo, amplitud, size)}")
```

```python
#Establecemos los parametros mencionados arriba para la Distribucion Uniforme
# Parámetros de la distribución
minimo = 0
maximo = 10
amplitud = maximo - minimo

#Creamos el eje x con el minimo y maximo, ademas de 100 valores aleatorios dentro de esos valores
x = np.linspace(minimo, amplitud, 100)

#Definimos las funciones de densidad y acumulada
pdf = dens_uniforme(x, minimo, amplitud)
cdf = acum_uniforme(x, minimo, amplitud)

#Definimos la funcion percentil, fijemos un percentil del 60%
percentil = 0.60
ppf = percen_uniforme(percentil, minimo, amplitud)

#Eje x para ppf (funcion percentil) por ser una funcion continua
q = np.linspace(0.01, 0.99, 100)
ppf_curva = percen_uniforme(q, minimo, amplitud)

#Creando graficas

fig, axis = plt.subplots(3, 1, figsize=(6, 8))

#Grafica pdf - Densidad de Probabilidad
axis[0].plot(x, pdf, "r-")
axis[0].set_title("PDF - Densidad de Probabilidad - Distribucion Uniforme")

#Grafica cdf - Distribucion Acumulada
axis[1].plot(x, cdf, "b-")
axis[1].set_title("CDF - Distribución Acumulada - Distribucion Uniforme")

#Grafica ppf - Funcion Percentil
axis[2].plot(q, ppf_curva, "g-")
axis[2].set_title("PPF - Funcion Percentil - Distribucion Uniforme")
axis[2].axvline(percentil, color="blue", linestyle="--")
axis[2].axhline(ppf, color="gray", linestyle="--")
axis[2].plot(percentil, ppf, "go", label="Percentil 60%")
axis[2].legend()

plt.tight_layout()
plt.show()
print()
print(f'El valor donde se encuentra el 60% de los datos es {ppf}')
```
